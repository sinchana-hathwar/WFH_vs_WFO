```python
#Defining a function to split the year value and store it in a new column called year
def year_split(current):
    #Extracting the first four digits that represent the year
    current_year = int(current) // 100
    return current_year

#Defining a function to split the week value and store it in a new column called week    
def week_split(current):
    #Extracting the last two digits that represent the week
    current_week = int(current) % 100
    return current_week
```

```python
#Applying the year_split function to get the year values into the year column
negative_data['year'] = negative_data['year_week'].apply(year_split)
#Applying the week_split function to get the week values into the week column
negative_data['week'] = negative_data['year_week'].apply(week_split)
```
